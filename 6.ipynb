ii)	Measure Degree Centrality

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from pyvis.network import Network


# Convert adjacency matrix to graph
adjacent_matrix_graph = nx.Graph()
network = Network(notebook=True, directed =True)
network.add_nodes(df.index.to_list())
adjacent_matrix_graph.add_nodes_from(df.index.to_list())
net.repulsion()
```

```python
for index, row in df.iterrows():
    for key, value in row.to_dict().items():
        if value == 1:
            network.add_edge(index, key)
            adjacent_matrix_graph.add_edge(index, key)
```

```python
# Degree Centrality
nx.degree_centrality(adjacent_matrix_graph)
```
