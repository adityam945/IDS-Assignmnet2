By looking at the data, we can see a lot of NaN values, and we can replace them with 0. We can also see that the row and column counts do not match. In an adjacency matrix, the number of columns and rows fit evenly. So let us perform some EDA to find out the root cause of the problem.

### Sorting
We start out first by sorting the rows and columns to aline the entries

```python
df = df.sort_index()
df = df.sort_index(axis=1)
```

We can see that there is a duplicate index entry so let us drop the second index entry that is found.

```python
df = df.reset_index().drop_duplicates(subset='index', keep='first').set_index('index')
```

```python
df
```

```python
df.index
```

We can see that there are 0 entries in the index. We can drop it from the index as it is not associated with any name.

```python
df = df.drop(index="0")
```

```python
valid_columns = set(df.index)
```

Now we are going to drop all the columns that are not present in the index and keeping only the once that match or exsist.

```python
df = df.drop(columns=[col for col in df if col not in valid_columns])
```

```python
df
```

We are now going to fill in NaN values as we can assume that if it's NaN, they are not friends. We can also cross-check in the matrix indexes to see if it's 0 or 1, then fill in the NaN value, but it would be the assumption that they are friends.
If A[i][j] = 1, then A[j][i] can also be one, but we will not be doing as we are assuming that A might be friends with B, but B might not be friendly with A.

```python
df = df.fillna(0)
```

ii)	Measure Degree Centrality

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from pyvis.network import Network


# Convert adjacency matrix to graph
G = nx.Graph()
net = Network(notebook=True, directed =True)
net.add_nodes(df.index.to_list())
G.add_nodes_from(df.index.to_list())
net.repulsion()
```

```python
for index, row in df.iterrows():
    for key, value in row.to_dict().items():
        if value == 1:
            net.add_edge(index, key)
            G.add_edge(index, key)
```

```python
net.show('nx.html')
```

```python
nx.degree_centrality(G)
```

Measure Closeness Centrality

```python
nx.closeness_centrality(G)
```
